### 7. Cloak images

To cloak your images, you need to specify the image path.
if you want to cloak multiply images, you can insert multiple paths with a comma (,) inbetween.

***Mode:*** Fawkes provides three different modes to use:
            Low, mid, high.
            The higher the mode, the stronger the cloak. \\
***Feature Extractor:*** Fawkes uses two feature extractors: 'extractor_0' and 'extractor_2' \\
'extractor_0' can be used in 'mid' and 'high' mode. \\
'extractor_2' can be used in all three modes. \\
Fawkes does not explain any difference between these two extractors. \\
***GPU:*** If GPU is not None, Fawkes tries to initialize the GPU. \\

Expected time per cloak(without GPU): \\
Low: ~ 1min \\
mid: < 5min \\
high: < 10min \\

Expected time per cloak(with T4 GPU): \\
Low: < 15 seconds \\
mid: < 1.5 min \\
high: < 3 min \\

The cloaked images get uploaded into the same directory as the original images.


```python
!python3.8 -c "from fawkes.protection import Fawkes; \
image_paths = ['/content/drive/MyDrive/Fawkes_Images/demo/Fawkes_test_09.jpg', '/content/drive/MyDrive/Fawkes_Images/demo/Fawkes_test_13.jpg', '/content/drive/MyDrive/Fawkes_Images/demo/Fawkes_test_20.jpg']; \
protector = Fawkes(mode='low', feature_extractor='extractor_2', gpu="0", batch_size=1); \
protector.run_protection(image_paths, debug=True)"
```
